```python
import pandas as pd
import numpy as np
import math as m
import matplotlib.pyplot as plt
```

```python
bank_churners = pd.read_csv('BankChurners.csv', delimiter = ',')
```

```python
bank_churners.head(30)
```

```python
bank_churners.value_counts('Gender') # абсолютні
```

```python
bank_churners['Gender'].value_counts(normalize=True) # відносні
```

```python
bank_churners['Card_Category'].value_counts(ascending=True, normalize=True) # відносні
```

```python
bank_churners['Card_Category'].value_counts(ascending=True) # асболютні
```

```python
bcm = bank_churners['Customer_Age'].max()
bins = m.ceil((bcm-25)/10)
bins = np.arange(25, bcm, 10)
```

```python
bank_churners['Customer_Age'].value_counts(bins = bins) # асболютні
```

```python
bank_churners['Customer_Age'].value_counts(bins = bins, normalize=True) # відносні
```

```python
bank_churners.sort_values(by = ['Customer_Age'], ascending=False)
```

```python
bank_churners.sort_values(by = ['Total_Trans_Amt', 'Customer_Age'], ascending=True)
```

```python
plt.plot(bank_churners.groupby('Card_Category').size()) 
plt.xlabel("category")
plt.ylabel("count")
```

```python
bank_churners.groupby(pd.cut(bank_churners['Customer_Age'], bins))['Customer_Age'].sum().plot.pie() 
```

```python
plt.hist(bank_churners['Customer_Age'])
plt.xlabel("age")
plt.ylabel("count")
```

```python
pd.DataFrame(bank_churners.corrwith(bank_churners['Customer_Age'], method='pearson')) 
```
